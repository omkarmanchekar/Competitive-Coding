```python
A='12:03'
B='12:03'


def sol(A,B):
    A=[int(x) for x in A.split(':')]
    B=[int(x) for x in B.split(':')]
    hours=-1
    start_mintues=[00,15,30,45]
    end_mintues=[45,30,15,00]
    waste_quaters=[]

    if(A[0]-B[0] == 0 and A[1]-B[1] == 0):
        return 0
        
    if(A[0] <= B[0]):
        if(B[1] > 0):
            hours=B[0]-A[0]+1
        else:
            hours=B[0]-A[0]

    if(A[0]>B[0]):
        if(B[1] > 0):
            hours=(24-A[0])+(B[0])+1
        else:
            hours=(24-A[0])+(B[0])

    for i in range(len(start_mintues)):
        if(A[1] <= start_mintues[i]):
            waste_quaters.append(-i)
            break

    if(B[1]>0):
        for j in range(len(end_mintues)):
            if(B[1] >= end_mintues[j]):
                print("j",j)
                waste_quaters.append(-(j+1))
                break


    
    return hours*4+sum(waste_quaters)



sol(A,B)
```
